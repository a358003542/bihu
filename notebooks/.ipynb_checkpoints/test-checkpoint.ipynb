```python

import matplotlib.pyplot as plt

from matplotlib import rcParams

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'].insert(0, 'SimHei')  # 插入中文字体
# print(rcParams['font.sans-serif'])
rcParams['axes.unicode_minus'] = False  # 用来正常显示负号



import numpy as np

x = np.arange(6)
y = [2,2.1, 2.2, 2.1, 2.3, 2.4]
plt.ylabel("利润") # 设置y标签
plt.xlabel("月份") # 设置x标签
plt.xticks(x, ['七月', '八月', '九月', '十月', '十一月', '十二月']) # 设置x标记
plt.ylim(0, 2.8) # 设置y轴范围
plt.plot(y)
```
